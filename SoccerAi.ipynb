```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

import plotly.graph_objects as go

from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("/content/ITALY -Table 1.csv")

df = df.set_index('Team')

df
```

```python
Teams_Strength = df.loc[:,('Rating','GamePlyd','Goals','Possession%','Pass%','Shots pg','Shots OT pg','SuccDrib')]

Teams_Strength['GPG'] = (Teams_Strength['Goals'] / Teams_Strength['GamePlyd']).round(decimals=2) 

Teams_Strength['GPG_SOT_%'] = (Teams_Strength['GPG']/Teams_Strength['Shots OT pg']).round(decimals=2) 

Teams_Strength['SOT_%'] = (Teams_Strength['Shots OT pg']/Teams_Strength['Shots pg']).round(decimals=2) 

Teams_Strength['PP_Avg'] = ((Teams_Strength['Possession%']+Teams_Strength['Pass%'])-
                            (Teams_Strength['Possession%']+Teams_Strength['Pass%']).mean())

Teams_Strength['S_Avg'] = (Teams_Strength['Shots pg']-(Teams_Strength['Shots pg']).mean())

Teams_Strength['SOT_Avg'] = (Teams_Strength['Shots OT pg'] - ((Teams_Strength['Shots OT pg']).mean()))

Teams_Strength['Succ_Drib_Avg'] = (Teams_Strength['SuccDrib']-(Teams_Strength['SuccDrib'].mean()))



Teams_Opportunity = df.loc[:,('Rating','GamePlyd','Goals','AerialsWon','Inception','KeyLong','KeyShort','AccSP','AccLB')]

Teams_Opportunity['Aerial_Incep_Avg'] = ((Teams_Opportunity['AerialsWon']+Teams_Opportunity['Inception'])
                                        -(Teams_Opportunity['AerialsWon']).mean())

Teams_Opportunity['Key_Lo_Sh_Avg'] = ((Teams_Opportunity['KeyLong']+Teams_Opportunity['KeyShort'])
                                     -(Teams_Opportunity['KeyLong']+Teams_Opportunity['KeyShort']).mean())

Teams_Opportunity['Acc_LS_Avg'] = ((Teams_Opportunity['AccLB']+Teams_Opportunity['AccSP'])
                                  -(Teams_Opportunity['AccLB']+Teams_Opportunity['AccSP']).mean())



Teams_Weakness = df.loc[:,('Rating','GamePlyd','GoalCon','InAccSP','InAccLB','UnsuccDrib','Disposses','ArielLost')]

Teams_Weakness['In_LSB_Avg'] = ((Teams_Weakness['InAccLB']+Teams_Weakness['InAccSP'])
                               -(Teams_Weakness['InAccLB']+Teams_Weakness['InAccSP'].mean()))

Teams_Weakness['UnDrib_Dpos_AriL_Avg'] = ((Teams_Weakness['UnsuccDrib']+Teams_Weakness['Disposses']+Teams_Weakness['ArielLost'])
                                         -(Teams_Weakness['UnsuccDrib']+Teams_Weakness['Disposses']+Teams_Weakness['ArielLost']).mean())




Teams_Threat = df.loc[:,('Rating','GamePlyd','GoalCon','ShotsCon','DribbledPast','Fouls','Yellow','Red')]


Teams_Threat ['GC_PG'] = (Teams_Threat['GoalCon']/Teams_Threat['GamePlyd']).round(decimals=2)

Teams_Threat ['GC_%_SC'] = (Teams_Threat['GC_PG']/Teams_Threat['ShotsCon']).round(decimals=2)

Teams_Threat ['SCon_DriPast_Avg'] = ((Teams_Threat['ShotsCon']+Teams_Threat['DribbledPast'])
                                    -(Teams_Threat['ShotsCon']+Teams_Threat['DribbledPast']).mean())

Teams_Threat ['Foul_Y_R_Avg'] = ((Teams_Threat['Fouls']+Teams_Threat['Yellow']+Teams_Threat['Red'])
                                 -(Teams_Threat['Fouls']+Teams_Threat['Yellow']+Teams_Threat['Red']).mean())

```

```python
Teams_Strength
```

```python
Teams_Strength_Opportunity = pd.merge(Teams_Strength[['Rating','GamePlyd','Goals','GPG','GPG_SOT_%','SOT_%','PP_Avg','S_Avg','SOT_Avg','Succ_Drib_Avg']], Teams_Opportunity[['Aerial_Incep_Avg','Key_Lo_Sh_Avg','Acc_LS_Avg']] , left_index=True, right_index=True)
Teams_Strength_Opportunity
```

how about 1 teams strength and oppurtinutiy vs other teams weakness and treats and vise versa 

True Strenght of Team A = Team A Strength_Opportunity + Team B Threat_Weakness so does the True Strenght of Team B = Team B Strenght_Opportunity + Team A Threat_Weakness 

```python
Teams_Threat_Weakness = pd.merge(Teams_Threat[['Rating','GamePlyd','GoalCon','GC_PG','GC_%_SC','SCon_DriPast_Avg','Foul_Y_R_Avg']], Teams_Weakness[['In_LSB_Avg','UnDrib_Dpos_AriL_Avg']] , left_index=True, right_index=True)
Teams_Threat_Weakness
```

```python
Team_Data = pd.merge(Teams_Strength_Opportunity,Teams_Threat_Weakness,left_index=True, right_index=True)
```
